```python
sns.countplot(x= 'Gender', data = df)
```

sns.countplot(x='Gender', data=df): This creates a count plot for the Gender column from the DataFrame df, showing the number of occurrences of each gender category.



```python
ax = sns.countplot(x= 'Gender', data = df)
ax.bar_label(ax.containers[0])
plt.show()


# ax.bar_label(ax.containers[0]): This line adds bar labels to the count plot.
#The method ax.bar_label() places labels on the bars, displaying the count values above the bars.
#The argument ax.containers[0] refers to the first container of bars in the plot.
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

*df.groupby(['Gender'], as_index=False): Groups the DataFrame df by the Gender column, ensuring the Gender column remains in the final output (since as_index=False).*

**['Amount'].sum(): Aggregates (sums) the values in the Amount column for each gender group.**

**.sort_values(by='Amount', ascending=False): Sorts the resulting DataFrame in descending order based on the summed Amount values, so the highest total Amount appears first.**

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
ax=sns.barplot(x= 'Gender', y = 'Amount', data = sales_gen)
ax.bar_label(ax.containers[0])
ax.set_title('Amount spent by Gender')
plt.show()
```

```python
ax = sns.countplot(x= 'Age Group', data = df,hue='Gender')
for container in ax.containers:
    ax.bar_label(container)
plt.show()

#hue='Gender': The hue parameter in Seaborn allows you to split the data by a categorical variable—in this case, Gender.
#It adds a second level of categorization by displaying different colors for each value of the Gender column within the bars of each Age Group.
```

```python
# order by different states
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
ax = sns.barplot(x= 'State', y = 'Orders', data = sales_state)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
ax = sns.countplot(x = 'Marital_Status', data= df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
ax = sns.barplot(x= 'Marital_Status', y = 'Amount', hue = 'Gender', data = sales_state)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```
